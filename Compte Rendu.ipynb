### Ridge

```r
#conversion en matrice
x_train = as.matrix(x_train)
x_test = as.matrix(x_test)
y_train = as.matrix(y_train)
y_test = as.matrix(y_test)
```

```r
# Cross Validation for obtaining best value of lambda for ridge regression (alpha = 0)
# Fit ridge classifier
fit.ridge <- cv.glmnet(x_train, y_train, family = "gaussian", nfolds = 5, alpha = 0, standardize = TRUE,type.measure = "mse")
```

```r
# Checking best value of lambda
print(fit.ridge)
```

```r
# Storing best value of lambda 
best_lambda = fit.ridge$lambda.min
```

```r
#Cross-validation
lam <- cv.glmnet(x_train, y_train, family = "gaussian", nfolds = 5, alpha = 0, standardize = TRUE)
plot(lam)
```

```r
# Fitting Ridge Regression Model
ridge = glmnet(x_train, y_train, family = "gaussian",alpha = 0, lambda = best_lambda)

```

```r
summary(ridge)
```

```r
eval_results <- function(true, predicted, df) {
  SSE <- sum((predicted - true)^2)
  SST <- sum((true - mean(true))^2)
  R_SQUARE <- 1 - SSE / SST
  RMSE = sqrt(SSE/nrow(df))

  
data.frame(
  RMSE = RMSE,
  R_SQUARE = R_SQUARE
)
}
```

```r
# Checking its performance on train data set
predict_train <- ridge %>% predict(x_train)
data.frame( R2 = R2(predict_train, y_train),
            RMSE = RMSE(predict_train, y_train),
            MAE = MAE(predict_train, y_train))
```

```r
#tester le modèle
predict_test <- ridge %>% predict(x_test)
data.frame( R2 = R2(predict_test, y_test),
            RMSE = RMSE(predict_test, y_test),
            MAE = MAE(predict_test, y_test))
```

# **Lasso**

```r
# Cross Validation for obtaining best value of lambda for lasso regression (alpha = 1)
cv_for_best_lambda <-  cv.glmnet(x_train, y_train, family = "gaussian", alpha = 1,standardize = TRUE ,type.measure = "mse")
```

```r
summary(cv_for_best_lambda)
```

```r
# Find best lambda 
lam_lasso <- cv.glmnet(x_train, y_train, family = "gaussian", nfolds = 5, alpha = 1, standardize = TRUE)
plot(lam_lasso)
```

```r
print(lam_lasso$lambda.min)
```

```r
# Checking best value of lambda
print(cv_for_best_lambda)
```

```r
# Storing best value of lambda in a R-object best_lambda
best_lambda = cv_for_best_lambda$lambda.min
```

```r
# Fitting Lasso Regression Model
lasso = glmnet(x_train, y_train, family = "gaussian", alpha = 1,standardize = TRUE , lambda = best_lambda)
```

```r
summary(lasso)
```

```r
#tester le modéle 
prediction_train <- predict(lasso, s = best_lambda, newx = x_train)
eval_results(y_train, prediction_train, df)
```

```r
#tester le modéle 
prediction_test <- predict(lasso , s = best_lambda, newx = x_test)
eval_results(y_test, prediction_test, df)
```

### elasticnet

```r
# Cross Validation for obtaining best value of alpha and lambda for elasticnet regression 

# Define Training Control
set.seed(123)
train.control <- trainControl(method = "repeatedcv", number = 10, repeats = 5 )
# Train the model
cv_for_best_value <- train(price ~ ., data = df, method="glmnet", trControl = train.control)
# Obtaining Best Value of alpha and lambda
cv_for_best_value$bestTune
```

la sortie ci-dessus donne la meilleure valeur de alpha et lambda. Nous utiliserons ces valeurs dans l’ajustement de régression Elasticnet.

```r
plot(cv_for_best_value, main = "Elastic Net Regression")
```

```r
# Fitting Elasticnet Regression Model
enet <- glmnet(x_train, y_train, alpha = 0.1, lambda = 515.7968 ,family = "gaussian")
```

```r
summary(enet)
```

```r
# Checking its performance on train data set
predict_train <- enet %>% predict(x_train)
data.frame( R2 = R2(predict_train, y_train),
            RMSE = RMSE(predict_train, y_train),
            MAE = MAE(predict_train, y_train))
```

```r
# Checking performance of Elasticnet regression on test data set
predict_test <- enet %>% predict(x_test)
data.frame( R2 = R2(predict_test, y_test),
            RMSE = RMSE(predict_test, y_test),
            MAE = MAE(predict_test, y_test))
```

```r
#extract the best model
get_best_result = function(caret_fit) {
  best = which(rownames(caret_fit$results) == rownames(caret_fit$bestTune))
  best_result = caret_fit$results[best, ]
  rownames(best_result) = NULL
  best_result
}
```
